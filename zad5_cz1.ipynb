```python
# Zdefiniowana funkcja process_age uzupełnia brakujące komórki w kolumnie wiek(wartość -0.5) oraz tworzy przedziały
def process_age(df,cut_points,label_names):
    df["Age"] = df["Age"].fillna(-0.5)
    df["Age_categories"] = pd.cut(df["Age"],cut_points,labels=label_names)
    return df

cut_points = [-1,0, 5, 12, 18, 35, 60, 100]
label_names = ["Missing", 'Infant', "Child", 'Teenager', "Young Adult", 'Adult', 'Senior']

# Wykonywane jest nadpisanie po zastosowaniu funkcji process_age
train = process_age(train,cut_points,label_names)
test = process_age(test,cut_points,label_names)

# Z wykresu tabeli przestawnej otrzymujemy informacje ze ok. 70% niemowląt zostało uratowanych
age_cat_pivot = train.pivot_table(index="Age_categories",values="Survived")
age_cat_pivot.plot.bar()
plt.show()
```

```python
# Wykonanie zliczenia ile osób przynależy do każdej z klas
train['Pclass'].value_counts()
```

```python
column_name = "Pclass"
df = train
dummies = pd.get_dummies(df[column_name],prefix=column_name)
dummies.head()
```

```python
# Stworzenie trzech kolumn odnoszących się do klasy w jakiej podróżowali pasażerowie. Dzięki temu otrzymujemy
# informacje w postacji 1/0
def create_dummies(df,column_name):
    dummies = pd.get_dummies(df[column_name],prefix=column_name)
    df = pd.concat([df,dummies],axis=1)
    return df

train = create_dummies(train,"Pclass")
test = create_dummies(test,"Pclass")
train.head()
```

```python
# Dokonujemy podziału jak wyżej dla kolumny Sex oraz Age_categories dla obu zestawów(train oraz test)
train = create_dummies(train,"Sex")
test = create_dummies(test,"Sex")
train = create_dummies(train,"Age_categories")
test = create_dummies(test,"Age_categories")
```

```python
train.head()
```

```python
# Tworzenie modelu, Logistic Regression, użyta zostanie biblioteka scikit-learn
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
# Columns jest parametrem x dla modelu - informacje na podstawie których mamy stworzyć model przewidujący wynik
# Kolumna Survived jest parametrem y dla modelu - informacja o wyniku
columns = ['Pclass_1', 'Pclass_2', 'Pclass_3', 'Sex_female', 'Sex_male',
       'Age_categories_Missing','Age_categories_Infant',
       'Age_categories_Child', 'Age_categories_Teenager',
       'Age_categories_Young Adult', 'Age_categories_Adult',
       'Age_categories_Senior']

# Wywołanie metody uczącej model
lr.fit(train[columns], train['Survived'])
```

```python
# Przewidywanie wyniku ufności dla próbek, odległość próbki od hiperpłaszczyzny
lr.decision_function(train[columns])
```

```python
# Współczynnik cech funkcji decyzyjnej 
lr.coef_
```

```python
holdout = test 
               
# Podział danych na dwie grupy, 80% danych przeznaczonych do nauki, 20% do testu
from sklearn.model_selection import train_test_split

columns = ['Pclass_1', 'Pclass_2', 'Pclass_3', 'Sex_female', 'Sex_male',
       'Age_categories_Missing','Age_categories_Infant',
       'Age_categories_Child', 'Age_categories_Teenager',
       'Age_categories_Young Adult', 'Age_categories_Adult',
       'Age_categories_Senior']

all_X = train[columns]
all_y = train['Survived']

# train_X oraz test_X dane na podstawie których uczymy oraz testujemy model
# train_y oraz test_y zawiera wyniki 
train_X, test_X, train_y, test_y = train_test_split(all_X, all_y, test_size=0.2,random_state=0)
```

```python
# Wyświetlenie ilu wierszy i kolumn używamy do nauki
train_X.shape
```

```python
# Wyświetlenie ilu wierszy i kolumn używamy do testowania
test_X.shape
```

```python
# Stworzenie modelu dla danych przeznaczonych do nauki
from sklearn.metrics import accuracy_score
lr = LogisticRegression()
lr.fit(train_X, train_y)

# Wpisanie do zmiennej predictions przewidzianych etykiet dla zbioru text_X
predictions = lr.predict(test_X)
# Sprawdzenie poprawności przewidzianych etykiet, wynik to ok. 81%
accuracy = accuracy_score(test_y, predictions)
accuracy
```

```python
# Wyświetlenie wyniku w tabeli
# Czytając poziomo otrzymujemy informacje o rzeczywistym stanie np. Survived = 96+14, Died 20+49
# Czytając pionowo otrzymujemy informacje  o przewidzianym wyniku np. Survived = 96 TP 20 FP, Died 14 FN 49 TN
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(test_y, predictions)
pd.DataFrame(conf_matrix, columns=['Survived', 'Died'], index=[['Survived', 'Died']])
```

```python
# Dzielimy dane na porcje K. Z tych porcji K, część zestawów (wszystkie poza jednym, K-1) jest używana do treningu, 
# podczas gdy pozostały zestaw służy do testowania. Algorytm jest szkolony i testowany K razy, za każdym razem,
# gdy nowy zestaw jest używany jako zestaw testowy, podczas gdy pozostałe zestawy są używane do treningu.

# Tworzenie pełnego modelu z użyciem all_X i all_y
from sklearn.model_selection import cross_val_score
import numpy as np

lr = LogisticRegression()
scores = cross_val_score(lr, all_X, all_y, cv=10)
np.mean(scores)
```
